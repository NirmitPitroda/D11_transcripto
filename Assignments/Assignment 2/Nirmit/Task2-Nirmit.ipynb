We are training a LinearRegression Classifier
Linear Regression plots a staright line based on the data, and then we make predictions based on that line. 
In order to find the line of best fit for the data, we use Least Squares method.
The line is fit to the data to make the sum of squared distance of each training point from the line the smallest.
Also known as making the Sum of Squared Residuals (SS(fit) ) the least. 
We can add as many parameters as we want.
Equations with more parameters will never make SS(Fit) worse than equations with fewer parameters.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import Normalizer, MinMaxScaler

df=pd.read_csv('Pokemon.csv')
df['CAS']=df['Attack']+df['Sp. Atk']+df['Speed']*0.5
X=df[['Total','HP','Defense','Sp. Def']]
y=df['CAS']
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.25, random_state=42)
scaler=MinMaxScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
model=LinearRegression()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
print(f"Mean Squared Error: {mean_squared_error(y_test,y_pred):.2f}")
```

```python
#Plotting the Predicted CAS vs Actual CAS Values
plt.figure(figsize=(10,6))
plt.title("Predicted CAS vs Actual CAS")
plt.xlabel("Actual CAS")
plt.ylabel("Predicted CAS")
plt.scatter(y_test.values, y_pred, marker="o",alpha=0.3)
plt.show()
```
